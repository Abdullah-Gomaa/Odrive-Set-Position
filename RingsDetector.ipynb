```python
%cd /mydrive/Rings_Dataset/v1rings/

import glob
import os
import shutil
import math

images = glob.glob("*.jpg")
n_images = len(images)
n_train = math.ceil(0.8 * n_images)
n_val = math.floor(0.2 * n_images)
lbls = []
for i in range(n_train):
  img = images[i]
  lbl = img.replace('.jpg', '.txt')
  shutil.move("/mydrive/datasets/coco128/" + img, "/mydrive/datasets/coco128/images/train/" + img )
  shutil.move("/mydrive/datasets/coco128/" + lbl, "/mydrive/datasets/coco128/labels/train/" + lbl )
  lbls.append(img.replace('.jpg', '.txt'))


for i in range(n_train, n_images):
  img = images[i]
  lbl = img.replace('.jpg', '.txt')
  shutil.move("/mydrive/datasets/coco128/" + img, "/mydrive/datasets/coco128/images/valid/" + img )
  shutil.move("/mydrive/datasets/coco128/" + lbl, "/mydrive/datasets/coco128/labels/valid/" + lbl )
  lbls.append(img.replace('.jpg', '.txt'))
```
